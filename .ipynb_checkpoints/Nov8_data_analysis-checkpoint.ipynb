### Analysis of last week's lab

```python
from scipy.integrate import simps
A = empty(len(ds_array))
A2 = empty(len(ds_array2))
s = np.where(logical_and(freq > 2.09, freq < 2.13))
for k in range(30):
    F = fft.fft(ds_array[k])
    F2 = fft.fft(ds_array2[k])
    freq = fft.fftfreq(len(F), .1)
    freq2 = fft.fftfreq(len(F2), .1)
    A[k] = simps(abs(F[s]), freq[s])
    A2[k] = simps(abs(F2[s]), freq2[s])
```

```python
scatter(pol_t, A)  
xlabel('Polarization Time (ms)')
ylabel('Area under peak')
```

```python
scatter(pol_t, A2)  
xlabel('Polarization Time (ms)')
ylabel('Area under peak')
```

```python
from scipy.stats import sem
A_avg = mean(vstack((A, A2)), axis=0)
A_err = sem(vstack((A, A2)), axis=0)

scatter(pol_t, A_avg)
errorbar(pol_t, A_avg, yerr=A_err, linestyle="None")
xlabel('Polarization Time (ms)')
ylabel('Area under peak')
```

```python
def exp_model(t, T1, M0, d):
    return M0 * (1-exp(-t/T1)) + d
```

```python
from scipy.optimize import curve_fit
avg_popt, avg_pcov = curve_fit(exp_model, pol_t, A_avg, sigma=A_err, p0=[3000, 5000, 0])

scatter(pol_t, A_avg)
errorbar(pol_t, A_avg, yerr=A_err, linestyle="None")
plot(pol_t, exp_model(pol_t, avg_popt[0], avg_popt[1], avg_popt[2]))
xlabel('Polarization time (ms)')
ylabel('Signal Strength')
perr = np.sqrt(np.diag(avg_pcov))
avg_popt, avg_pcov, perr
```

```python
residuals = A_avg - exp_model(pol_t, avg_popt[0], avg_popt[1], avg_popt[2])
scatter(pol_t, residuals)
errorbar(pol_t, residuals, yerr=A_err, linestyle="None")
hlines(0, 0, 10000, colors='r')
title('Residuals')
```

```python
expected = exp_model(pol_t, avg_popt[0], avg_popt[1], avg_popt[2])
1/(size(pol_t)-3) * np.sum((A_avg - expected)**2 / A_err**2)
```

I find $T_1=2.76 \pm 0.098s$. My fit looks good with a random scatter of points around 0 in the residuals and a reduced $\chi^2$ near 1. 

I improved our model from last time (Nov. 1, 2017) by adding a vertical displacement, $d$, to the model. This allows for the fact that at 0 polarization time, the integral around the peak area is not 0 due to noise. This improved our fit and I went back to change the model fits from last lab. It is interesting to note that using the same data, but different analysis techniques, I get different $T_1$ measurements. For now note that this week's analysis produced a much better fit to the model than last week and this week's method involves less human input as the peak range is the same each time.

### How do the peak limits affect the $T_1$ obtained from the fit?

```python
for k in range(30):
    F = fft.fft(ds_array[k])
    freq = fft.fftfreq(len(F), .1)
    plot(freq, abs(F))
xlim([2.08, 2.18])
xlabel('Frequency (kHz)')
ylabel('Spectrum')
```

```python
center = 2.115
deltas = arange(.015, .031, .001)
A_avgs = zeros((len(deltas), len(pol_t)))
A_errs = zeros((len(deltas), len(pol_t)))
popts = zeros((len(deltas), 3))
T1s = zeros(len(deltas))
T1_errs = empty(len(deltas))
chi2s = empty(len(deltas))
counter = 1
for m in range(len(deltas)):
    A = empty(len(ds_array))
    A2 = empty(len(ds_array2))
    s = np.where(logical_and(freq > center-deltas[m], freq < center+deltas[m]))
    for k in range(30):
        F = fft.fft(ds_array[k])
        F2 = fft.fft(ds_array2[k])
        freq = fft.fftfreq(len(F), .1)
        freq2 = fft.fftfreq(len(F2), .1)
        A[k] = simps(abs(F[s]), freq[s])
        A2[k] = simps(abs(F2[s]), freq2[s])
    avg = mean(vstack((A, A2)), axis=0)
    err = sem(vstack((A, A2)), axis=0)
    popt, pcov = curve_fit(exp_model, pol_t, avg, sigma=err,
                                   p0=[2800, 3500+1500*counter/len(deltas), 146])
    A_avgs[m] = avg
    A_errs[m] = err
    popts[m] = popt
    perr = sqrt(np.diag(pcov))
    T1s[m] = popt[0]
    T1_errs[m] = perr[0]
    counter= counter+1
    expected = exp_model(pol_t, *popt)
    chi2s[m] = 1/(size(pol_t)-3) * np.sum((avg - expected)**2 / err**2)
```

```python
s = chi2s < 9  # do not include models that did not fit well
scatter(deltas[s], T1s[s])
errorbar(deltas[s], T1s[s], yerr=T1_errs[s], linestyle="none")
xlabel('Peak range')
ylabel('$T_1$')
std(T1s[s])
```

```python
for k in range(30):
    F = fft.fft(ds_array[k])
    freq = fft.fftfreq(len(F), .1)
    plot(freq, abs(F))
for delta in deltas:
    axvline(center-delta, color='r', linestyle=':')
    axvline(center+delta, color='r', linestyle=':')
xlim([2.06, 2.175])
xlabel('Frequency (kHz)')
ylabel('Spectrum')
```

```python
for k in range(len(A_avgs)):
    figure()
    scatter(pol_t, A_avgs[k])
    errorbar(pol_t, A_avgs[k], yerr=A_errs[k], linestyle='none')
    plot(pol_t, exp_model(pol_t, *popts[k]))
    print(popts[k])
```
